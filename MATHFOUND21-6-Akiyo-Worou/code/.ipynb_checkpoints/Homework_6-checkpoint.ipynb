```python
# importation of libraries
import torch
import glob
import os
from torch.utils.data import Dataset
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Task 1 - Datasets, Preprocessing and Data loading

```python
class SRDataset(Dataset):
    def __init__(self,path):
        self.images_filenames=glob.glob( os.path.join(path, '*.png')   )
    def __len__(self):
        return len(self.images_filenames)
    def __getitem__(self,index):
        im=torchvision.io.read_image(self.images_filenames[index])
        im=(1/255)*im
        hr_im=transforms.Compose([
        transforms.RandomCrop((64,64)),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2)
        ])(im)
        lr_im = transforms.Resize(32,interpolation=transforms.InterpolationMode.BILINEAR)(hr_im)  # low resolution image
        return lr_im, hr_im

      
```

```python
# test the class
train_path='train'
train_dataset = SRDataset(train_path)
# Assuming the train_dataset object is properly initialized
# and that the __getitem__ returns a tuple
train_dataloader = torch.utils.data.dataloader.DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True,
    num_workers=2,
    drop_last=True,
    pin_memory=True)
print(f" * Dataset contains {len(train_dataset)} image(s).")
for _, batch in enumerate(train_dataloader, 0):
    lr_image, hr_image = batch
    #lr_image=lr_image[0, ...].mul(255).byte()
    #hr_image=hr_image[0, ...].mul(255).byte()
    #print(lr_image.shape)
    #print(lr_image[0, ...].mul(255).byte().shape)
    torchvision.io.write_png(lr_image[0, ...].mul(255).byte(), "lr_image.png")
    torchvision.io.write_png(hr_image[0, ...].mul(255).byte(), "hr_image.png")
    break # we deliberately break after one batch as this is just a test

# evaluation 
eval_path='eval'

eval_dataset = SRDataset(eval_path)
eval_dataloader = torch.utils.data.dataloader.DataLoader(
        eval_dataset)
```

## Task 2 - Create the Model

```python
class BasicSRModel(torch.nn.Module):
    def __init__(self,interpolation='bilinear'):
        super().__init__()
        self.interpolation=interpolation
        self.first_conv=torch.nn.Conv2d(3,64,3,stride=1, padding=1)

        self.block_1=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_2=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_3=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_4=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_5=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_6=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_7=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_8=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_9=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_10=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])

        self.last_conv=torch.nn.Conv2d(64,3,3,stride=1, padding=1)

    def forward(self,x):
        x = torch.nn.functional.interpolate(x,scale_factor=(2,2),mode=self.interpolation)
        x=self.first_conv(x)
        x=self.block_1(x)
        x=self.block_2(x)
        x=self.block_3(x)
        x=self.block_4(x)
        x=self.block_5(x)
        x=self.block_6(x)
        x=self.block_7(x)
        x=self.block_8(x)
        x=self.block_9(x)
        x=self.block_10(x)
        x=self.last_conv(x)

        return x
```

## Task 3 - Implement the Training Loop

```python
model=BasicSRModel()

saving_path='basicSR_0.pth'

number_of_epoch = 120 
```

```python
def train_model(model,saving_path,learning_rate=1e-4):
    

    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()),
    lr=learning_rate)

    loss_function = torch.nn.L1Loss()

    Loss_list=[]
    for epoch in range(number_of_epoch):
        Global_loss=0.0
        for i, batch in enumerate(train_dataloader):
            low_res, high_res = batch
            # reset the gradient
            optimizer.zero_grad()
            # forward pass through the model
            high_res_prediction = model(low_res)
            # compute the loss
            loss = loss_function(high_res_prediction, high_res)
            # adding up
            Global_loss+=loss.item()
            # backpropagation
            loss.backward()
            # update the model parameters
            optimizer.step()
        # saving state
        torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': Loss_list}, saving_path)
        # log the metrics, images, etc
        Global_loss/=len(train_dataloader)
        Loss_list.append(Global_loss)
        print('epoch : ', epoch,'-------------------------------------> loss : ',Global_loss)
    plt.plot(Loss_list)
```

```python
# New training loop


def train_and_evaluate(model,saving_path,learning_rate=1e-4,train_dataloader=train_dataloader,eval_dataloader=eval_dataloader):

    optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, model.parameters()),
lr=learning_rate)
    loss_function = torch.nn.L1Loss()
    L1_list=[]
    PSNR_list=[]
    SSIM_list=[]
    Loss_list=[]

    for epoch in range(number_of_epoch):
        Global_loss=0.0
        for i, batch in enumerate(train_dataloader):
            low_res, high_res = batch
            # reset the gradient
            optimizer.zero_grad()
            # forward pass through the model
            high_res_prediction = model(low_res)
            # compute the loss
            loss = loss_function(high_res_prediction, high_res)
            # adding up
            Global_loss+=loss.item()
            # backpropagation
            loss.backward()
            # update the model parameters
            optimizer.step()
        
        # log the metrics, images, etc
        Global_loss/=len(train_dataloader)
        Loss_list.append(Global_loss)
        print('epoch : ', epoch,'-------------------------------------> loss : ',Global_loss)

        with torch.no_grad():
            n=0
            L1_loss=0;PSNR_loss=0;SSIM_loss=0
            for  _, batch in enumerate(eval_dataloader):
                low_res, high_res = batch
                # forward pass through the model
                high_res_prediction = model(low_res.float())

                L1_loss+=loss_function(high_res_prediction, high_res)
                PSNR_loss+=PSNR(high_res_prediction, high_res)
                SSIM_loss+=ssim(high_res_prediction, high_res)

                n+=1
            L1_list.append(L1_loss/n)
            PSNR_list.append(PSNR_loss/n)
            SSIM_list.append(SSIM_loss/n)
    # saving state
        torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': Loss_list,
                'L1_eval': L1_list,
                'PSNR_eval': PSNR_list,
                'SSIM_eval':SSIM_list}, saving_path)
    # Metrics plot
    
    plt.plot(Loss_list)
    plt.title('Training Loss')

    fig = plt.figure(figsize=(20,15))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)

    ax2 = fig.add_subplot(2, 2, 1)
    ax2.plot(L1_list)
    ax2.set_title('Evaluation error with L1 metrics')

    ax3 = fig.add_subplot(2, 2, 2)
    ax3.plot(PSNR_list)
    ax3.set_title('Evaluation error with PSNR metrics')

    ax4 = fig.add_subplot(2, 2, 3)
    ax4.plot(SSIM_list)
    ax4.set_title('Evaluation error with SSIM metrics')

    ax5 = fig.add_subplot(2, 2, 4)
    ax5.plot( 0.5*(np.array(SSIM_list) +np.array(PSNR_list) ) )
    ax5.set_title('Evaluation error with average PSNR and SSIM metrics')

    plt.show()
    
```

```python
class ResidualSRModel(torch.nn.Module):
    def __init__(self,interpolation="bilinear"):
        super().__init__()
        self.interpolation=interpolation
        self.first_conv=torch.nn.Conv2d(3,64,3,stride=1, padding=1)

        self.block_1=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_2=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_3=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_4=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_5=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_6=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_7=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_8=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_9=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])
        self.block_10=torch.nn.Sequential(*[torch.nn.Conv2d(64,64,3,stride=1, padding=1),torch.nn.LeakyReLU()])

        self.last_conv=torch.nn.Conv2d(64,3,3,stride=1, padding=1)

    def forward(self,x):
        x = torch.nn.functional.interpolate(x,scale_factor=(2,2),mode=self.interpolation)
        residual=x
        x=self.first_conv(x)
        x=self.block_1(x)
        x=self.block_2(x)
        x=self.block_3(x)
        x=self.block_4(x)
        x=self.block_5(x)
        x=self.block_6(x)
        x=self.block_7(x)
        x=self.block_8(x)
        x=self.block_9(x)
        x=self.block_10(x)
        x=self.last_conv(x)
        x+=residual
        return x
```

```python
class Dense_block(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.block=torch.nn.Sequential(*[torch.nn.Conv2d(32,32,3,stride=1,padding=1),
                                                      torch.nn.BatchNorm2d(32,0.8),
                                                      torch.nn.PReLU(),
                                                      torch.nn.Conv2d(32,32,3,stride=1,padding=1),
                                                      torch.nn.BatchNorm2d(32,.8)])
    def  forward(self,x):
        x=x+self.block(x)
        return x
```

```python
class Generator(torch.nn.Module):
    def __init__(self):
        super().__init__()
        
        self.first_conv=torch.nn.Sequential(*[torch.nn.Conv2d(3,32,9,stride=1,padding=4),torch.nn.PReLU()])

        self.residual_blocks=[]
        # Blocks
        for i in range(16):
            self.residual_blocks.append(Dense_block() )

        self.endblock_conv=torch.nn.Sequential(*[torch.nn.Conv2d(32,32,3,stride=1,padding=1),torch.nn.BatchNorm2d(32,.8)])

        self.upsampling_conv=torch.nn.Sequential(*[torch.nn.Conv2d(32, 128, 3,stride=1,padding=1),
                                                    torch.nn.BatchNorm2d(128),
                                                    torch.nn.PixelShuffle(2),
                                                    torch.nn.PReLU()])
       
        
        self.last_conv=torch.nn.Sequential(*[torch.nn.Conv2d(32, 3, kernel_size=9, stride=1, padding=4), torch.nn.Tanh()])

    def forward(self, x):
        
        output_1=self.first_conv(x)
        output_2=torch.clone(output_1)
        for i in range(16):
            output_2=self.residual_blocks[i](output_2)+output_2
        output_3=self.endblock_conv(output_2)+output_1
        output_4=self.upsampling_conv(output_3)
        
        #print(x.size(),output_1.size(),output_2.size(),output_3.size(),output_4.size())
        
        output = self.last_conv(output_4)
        #print(output.size())
        return output


        
```

```python
class Discriminator(torch.nn.Module):
    def __init__(self):
        super().__init__()

        self.blocks=[]
        
        L=[3,32,32,64,64,128,128,256,256]
        for i in range(8):
            if i ==0:
                self.blocks.append(torch.nn.Sequential(*[torch.nn.Conv2d(L[i],L[i+1],3,stride= 1 ,padding=1),
                                                         torch.nn.LeakyReLU(0.2, inplace=True)]))
            else:
                self.blocks.append( torch.nn.Sequential(*[torch.nn.Conv2d(L[i],L[i+1],3,stride= (i%2)+1 ,padding=1),
                                                      torch.nn.BatchNorm2d(L[i+1]),
                                                        torch.nn.LeakyReLU(0.2, inplace=True)  ]))
        
        self.last_block= torch.nn.Sequential(*[torch.nn.Linear(4,512),
                                torch.nn.LeakyReLU(),torch.nn.Linear(512,1),
                                torch.nn.Sigmoid()])  
        #self.last_block = torch.nn.Conv2d(256,3,3,stride = 1, padding=1)

    def forward(self,x):
        for i in range(8):
            #print(x.size())
            x=self.blocks[i](x)
        #print(x.size())
        x=self.last_block(x)
        #print(x.size())
        return x
```

```python
# Model training
saving_path_srgan='SRGAN_model.pth'
number_of_epoch=120
def SRGAN_training(generator,discriminator,saving_path_srgan, G_learning_rate=1e-4,D_learning_rate=1e-4,train_dataloader=train_dataloader,eval_dataloader=eval_dataloader):
    
    # optimizers 
    generator_optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, generator.parameters()),
lr=G_learning_rate)
    discriminator_optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, discriminator.parameters()),
lr=D_learning_rate)
    
    loss_function = torch.nn.L1Loss()
    MSEloss_function=torch.nn.MSELoss()
    
    G_Loss_list=[]
    D_Loss_list=[]

    for epoch in range(number_of_epoch):
        # Losses initialization
        G_loss=0.0
        D_loss=0.0

        for i, batch in enumerate(train_dataloader):
            low_res, high_res = batch
            generator.train()
            discriminator.train()

            
            real = torch.autograd.Variable(torch.Tensor(np.ones((4,256,4,1))), requires_grad=False)
            fake = torch.autograd.Variable(torch.Tensor(np.zeros((4,256,4,1))), requires_grad=False)
            # reset the generator gradient
            generator_optimizer.zero_grad()
            # forward pass through the generator model
            high_res_generator = generator(low_res)
            # compute the loss
            # adversarial loss
            adv_loss=MSEloss_function(discriminator(high_res_generator),real)

            generator_loss = loss_function(high_res_generator, high_res.detach()) + 1e-3*adv_loss
            # adding up
            G_loss+=generator_loss.item()
            # backpropagation
            generator_loss.backward()
            # update the model parameters
            generator_optimizer.step()


            # reset the discriminator gradient
            discriminator_optimizer.zero_grad()

            # forward pass through the discriminator model
            #print(high_res.size())
            high_res_disc_real = discriminator(high_res)
            high_res_disc_fake = discriminator(high_res_generator.detach())
            

            # compute the loss
            real_pred_loss = MSEloss_function(high_res_disc_real, real)
            fake_pred_loss = MSEloss_function(high_res_disc_fake,fake)
            discriminator_loss=0.5*( real_pred_loss+ fake_pred_loss)
            # adding up
            D_loss+=discriminator_loss.item()
            # backpropagation
            discriminator_loss.backward()
            # update the model parameters
            discriminator_optimizer.step()

        # log the metrics, images, etc
        G_loss/=len(train_dataloader)
        D_loss/=len(train_dataloader)
        G_Loss_list.append(G_loss)
        D_Loss_list.append(D_loss)

        print('epoch : ', epoch,'-------------------------------------> generator loss : ',
              G_loss,' discriminator loss : ', D_loss)

    # saving state
        torch.save({
                'epoch': epoch,
                'generator_state_dict': generator.state_dict(),
                'discriminator_state_dict': discriminator.state_dict(),
                'G_state_dict': generator_optimizer.state_dict(),
                'D_state_dict': discriminator_optimizer.state_dict(),
                'generator_loss': G_Loss_list,
                'discriminator_loss': D_Loss_list}, saving_path_srgan)
    # Metrics plot
    
    plt.plot(G_Loss_list)
    plt.title('Generator Training Loss')
    plt.figure()
    plt.plot(D_Loss_list)
    plt.title('Discriminator Training Loss')

    plt.show()
```

```python
SRGAN_training(generator,discriminator,saving_path_srgan)
```
