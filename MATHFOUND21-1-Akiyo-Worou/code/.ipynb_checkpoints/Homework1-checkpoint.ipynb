##### PART 1: RANSAC FOR CIRCLE FITTING

###### 1-1 Data generation

```python
# Constants used for the data generation
x_c=0; y_c=0; R=5
noise=0.1; r=0.2; N=100
```

```python
def generate_data_circle(N,r,tau):
    
    # getting the number of inliers and outliers
    nb_outliers=int(N*r)
    nb_inliers=N-nb_outliers
    inliers=[];outliers=[]
    # Adding inliers to the dataset
    while len(inliers)<nb_inliers:
        x=20*random()-10
        y=20*random()-10
        v=(x-x_c)**2+(y-y_c)**2
        if v<=(R*(1+noise))**2 and v>=(R*(1-noise))**2:
            inliers.append([x,y])
            
    #Adding outliers to the dataset
    while len(outliers)<nb_outliers:
        x=20*random()-10
        y=20*random()-10
        v=(x-x_c)**2+(y-y_c)**2
        if v>(R*(1+noise))**2 or v<(R*(1-noise))**2:
            outliers.append([x,y])
            
    inliers=np.array(inliers)
    outliers=np.array(outliers)
    
    #creating the data set
    data=np.concatenate((inliers,outliers))
    return data,inliers,outliers
```

```python
# Checking the result of the previous fonction

(data,inliers,outliers)=generate_data_circle(100,r,0.1)
plt.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7)
plt.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7)


#plot the circle obtained
Theta=np.linspace(0,2*np.pi,1000)
X=[R*np.cos(theta)+x_c for theta in Theta]
Y=[R*np.sin(theta)+y_c for theta in Theta]
plt.plot(X,Y,color='green')
plt.axis("equal")
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()
```

###### 1-2 RANSAC

The number of iteration M is given by $$M=\frac{log(1-p)}{log(1-(1-r)^s)}$$
With $p$ the probability that at least one sample has no outliers, $s$ a sample size and $r$ the outlier ratio<br>
$p=0.99$ and $s=3$ 

```python
# data used
r_list=[0.05,0.2,0.3,0.7]
seed(9001)
data_list=[generate_data_circle(N,r,0.1) for r in r_list]
```

```python
def OneRansac(data,r):
    M=int(np.log10(1-0.99)/np.log10(1-(1-r)**3))+1
    Ransac_nb_inliers=0
    
    for i in range(M):
        temp_threePoints=np.array(sample(list(data), 3))
        
        # the center of gravity is the center of the circle
        gravity=np.mean(temp_threePoints,axis=0)
        
        # Detecting the inliners and outliers
        # Compute the ratio of the distance from a point in the center to the radius of the circle
        is_an_inlier=[ 0.9<=np.linalg.norm(data[i]-gravity)/np.linalg.norm(gravity-temp_threePoints[0])<=1.1 
                     for i in range(len(data)) ]
        
        temp_inliers=np.sum(is_an_inlier)
        if temp_inliers>Ransac_nb_inliers:
            threePoints=temp_threePoints
            Ransac_nb_inliers=temp_inliers
    # getting the final center of gravity and the radius
    center=np.mean(threePoints,axis=0)
    radius=np.linalg.norm(center-threePoints[0])
    return [Ransac_nb_inliers,center, radius]
```

```python
def Ransac(data,r):
    start_time=time.process_time()
    
    # Applying the Ransac 10000 times 
    L=np.array([OneRansac(data,r) for n in range(1000)])
    inliers_dist=L[:,0]
    
    # The center and the radius of the circle with the highest number of inliers
    i=np.argmax(inliers_dist)
    center=L[i,1]
    radius=L[i,2]
    inliers=[]
    outliers=[]
    for datum in data:
        if 0.9<=np.linalg.norm(datum-center)/radius<=1.1:
            inliers.append(datum)
        else:
            outliers.append(datum)
    inliers=np.array(inliers)
    outliers=np.array(outliers)
    
    duration=time.process_time()-start_time
    return [inliers_dist,inliers,outliers,center,radius,duration]
```

```python
def Ransac_results_plot(data_list):
    m=len(data_list)
    fig = plt.figure(figsize=(20,18))
    fig.subplots_adjust(hspace=0.7, wspace=0.7)
    for i in range(m):
        (data,synthesized_inliers,_)=data_list[i]
        ransac=Ransac(data,r_list[i])
        
    # ransac distribution
        ax=fig.add_subplot(3,m,i+1)
        ax.hist(ransac[0],bins=int(180/4))
        ax.set_xlabel('Nb of detected inliers')
        ax.set_ylabel('Nb of experiments')
        
    #ransac detection
        ax1=fig.add_subplot(3,m,i+1+m)
        # plot the inliers and outliers
        inliers=ransac[1]
        outliers=ransac[2]
        inlier_ratio_label='Ransac inliers with '+ str(len(ransac[1])/len(synthesized_inliers) ) + ' ratio'
        ax1.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7,label=inlier_ratio_label )
        ax1.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7,label='Ransac outliers')


        #plot the real circle
        Theta=np.linspace(0,2*np.pi,1000)
        X=[R*np.cos(theta)+x_c for theta in Theta]
        Y=[R*np.sin(theta)+y_c for theta in Theta]
        ax1.plot(X,Y,color='green',label='Synth. model')
        
        # plot the Ransac circle
        radius=ransac[4]
        center=ransac[3]
        X1=[radius*np.cos(theta)+center[0] for theta in Theta]
        Y1=[radius*np.sin(theta)+center[1] for theta in Theta]
        ransac_time_label='Ransac model run for ' + str(ransac[5]) +' seconds'
        ax1.plot(X1,Y1,color='black',label=ransac_time_label)
        ax1.axis("equal")
        ax1.title.set_text("Results for the ratio r = "+ str(int(r_list[i]*100)) + " %")
        ax1.legend(loc="lower center",bbox_to_anchor=(0.5, -0.5))
```

###### 1.3 Exhaustive Search

```python
def exhaustive_search(data):
    start_time=time.process_time()
    
    # getting all the combinations
    exhaustive_list=list(combinations(data, 3))
    nb_inliers=0
    for l in exhaustive_list:
        temp_threePoints=l
        
        # the center of gravity is the center of the circle
        gravity=np.mean(temp_threePoints,axis=0)
        
        
        # Detecting the inliners and outliers
        # Compute the ratio of the distance from a point in the center to the radius of the circle
        is_an_inlier=[ 0.9<=np.linalg.norm(data[i]-gravity)/np.linalg.norm(gravity-temp_threePoints[0])<=1.1 
                     for i in range(len(data)) ]
        
        temp_inliers=np.sum(is_an_inlier)
        if temp_inliers>nb_inliers:
            threePoints=temp_threePoints
            nb_inliers=temp_inliers
            
    # getting the final center of gravity and the radius
    center=np.mean(threePoints,axis=0)
    radius=np.linalg.norm(center-threePoints[0])
    
    # getting the inliers and outliers
    inliers=[]
    outliers=[]
    for datum in data:
        if 0.9<=np.linalg.norm(datum-center)/radius<=1.1:
            inliers.append(datum)
        else:
            outliers.append(datum)
    inliers=np.array(inliers)
    outliers=np.array(outliers)
    
    duration=time.process_time()-start_time
    return [inliers,outliers,center, radius,duration]
```

```python
def exhaustive_results_plot(data_list):
    m=len(data_list)
    fig = plt.figure(figsize=(20,20))
    fig.subplots_adjust(hspace=0.7, wspace=0.7)
    for i in range(m):
        (data,synthesized_inliers,_)=data_list[i]
        exhaustiveSearch=exhaustive_search(data)

        
    # exhaustive search
        ax2=fig.add_subplot(3,m,i+1)
        # plot the inliers and outliers
        inliers=exhaustiveSearch[0]
        outliers=exhaustiveSearch[1]
        inlier_ratio_label='Exhaustive Search inliers with '+ str(len(exhaustiveSearch[0])/len(synthesized_inliers) ) + ' ratio'
        ax2.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7,label=inlier_ratio_label)
        ax2.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7,label='Exhaustive Search outliers')


        #plot the real circle
        Theta=np.linspace(0,2*np.pi,1000)
        X=[R*np.cos(theta)+x_c for theta in Theta]
        Y=[R*np.sin(theta)+y_c for theta in Theta]
        ax2.plot(X,Y,color='green',label='Synth. model')
        
        # plot the Exhaustive circle
        radius=exhaustiveSearch[3]
        center=exhaustiveSearch[2]
        X1=[radius*np.cos(theta)+center[0] for theta in Theta]
        Y1=[radius*np.sin(theta)+center[1] for theta in Theta]
        exhaustive_time_label='Exhaustive Search model run for ' + str(exhaustiveSearch[4]) +' seconds'
        ax2.plot(X1,Y1,color='black',label=exhaustive_time_label)
        ax2.axis("equal")
        ax2.title.set_text("Results for the ratio r = "+ str(int(r_list[i]*100)) + " %")
        ax2.legend(loc="lower center",bbox_to_anchor=(0.5, -0.5))


        
        
        
```

###### IRLS with $L_1$ norm

The problem that we need to solve is the following: 
$$  \underset{{\theta}}{argmin}  \sum_{i=0}^{n} w_i(y_i-z_i\theta)^2=\underset{{\theta}}{argmin}[(y-z\theta)^Tw(y-z\theta)]$$
with $z_i=(x_i \hspace{0.25cm} 1)$, $\theta= (a \hspace{0.25cm} b)^T$, $z=(z_1,...,z_n)^T$ and $w_i=\frac{1}{2|y_i-z_i\theta|}$ $ \forall i \in {1,...,n}$ knowing that w is a diagonal matrix.
The gradient is then given by $2(z^Twz\theta-z^Twy)$ if we assume that w is not a function of $\theta$
We can then deduce the expression of $\theta$ at each step t+1
$$ \theta^{(t+1)}=(z^Tw^{(t)}z)^{-1}z^Tw^{(t)}y$$
with $w^{(t)}$ a function of $\theta^{(t)}$

```python
def IRLS(data,max_iter,epsilon=1e-5):
    y=data[:,1]
    z=np.array([ [data[:,0][i], 1] for i in range(len(data))])
    #choosing delta to avoid dividing by zero for the weigths computation
    delta=0.00001
    # initialisation of the weigths
    w=np.eye(len(data))
    A=np.dot(z.T,w)
    theta=np.dot(np.linalg.inv(np.dot(A,z)),np.dot(A,y))
    for t in range(max_iter):
        w_=np.diag([1/(2*max(delta,abs(y[i]-np.dot(z[i],theta)))) for i in range(len(data))])
        A_=np.dot(z.T,w_)
        theta_=np.dot(np.linalg.inv(np.dot(A,z)),np.dot(A,y))
        if np.linalg.norm(theta_-theta)<epsilon:
            return theta_
        theta=theta_
    return theta
```

```python
def IRLS_plot(data_list,max_iter):
    m=len(r_list)
    fig = plt.figure(figsize=(20,20))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    for i in range(m):
        (data,inliers,outliers)=data_list[i]
        
        #plot the real line
        ax2=fig.add_subplot(3,m,i+1)
        X=np.linspace(-10,10,1000)
        Y=[a*x+b for x in X]
        model_label='Synth model with a = '+ str(a)+ " and b = "+str(b)
        ax2.plot(X,Y,color='green',label=model_label)
        
        plt.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7,label='inliers')
        if r_list[i]!=0:
            plt.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7,label='outliers')
        
        #Plot IRLS line
        irls_line=IRLS(data,max_iter)
        a_,b_=irls_line[0],irls_line[1]
        
        X=np.linspace(-10,10,1000)
        Y=[a_*x+b_ for x in X]
        irls_label='IRLS model with a = '+ str(a_)+ " and b = "+str(b_)
        ax2.plot(X,Y,color='blue',label=irls_label)
        ax2.title.set_text("Results for the ratio r = "+ str(int(r_list[i]*100)) + " %")
        ax2.legend(loc="lower center",bbox_to_anchor=(0.5, -0.5))
        plt.xlim([-10, 10])
        plt.ylim([-10, 10])
```

###### Linear Programming with L1 and $L_{\infty}$ norms

By using Linear programming the aim is to minimize ${||y-ax-b||}_{p}$ for $L_p$ norm with $p = 1$ and $p= \infty$               
In linear programming, the problem can be written as follow: <br>
For $L_1$ norm the problem formulation is:  <br><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;$\underset{a,b}{min} \sum_{i=1}^n u_i$ 
&emsp;&emsp;&emsp;&emsp;subject to $\left\{
\begin{array}{l}
  u_i \geq y_i-ax-b \\
  u_i \geq -(y_i-ax-b)
\end{array}
\right.$<br><br>
For $L_{\infty}$ norm the problem formulation is:  <br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp; $\underset{a,b}{min}$ $u$
&emsp;&emsp;&emsp;&emsp;subject to $\left\{
\begin{array}{l}
  u \geq y_i-ax-b \\
  u \geq -(y_i-ax-b)
\end{array}
\right.$

###### Linear Programming with $L_1$ norm

```python
def linearProgL1(data):
    C=np.array([0,0]+[1 for i in range(N)]) # the vector considered here is a,b, u_1,....,u_n
    A_ub=np.zeros((2*N,N+2))
    b_ub=np.zeros(2*N)
    for i in range(N):
        A_ub[i,0]=-data[i,0]
        A_ub[i,1]=-1
        A_ub[i,i+2]=-1
        b_ub[i]=-data[i,1]
        
        A_ub[2*i+1,0]=data[i][0]
        A_ub[2*i+1,1]=1
        A_ub[2*i+1,i+2]=-1
        b_ub[2*i+1]=data[i,1]
        
    return linprog(C,A_ub=A_ub,b_ub=b_ub).x[:2]
    
    
```

```python
def linearProgL1_plot(data_list):
    m=len(r_list)
    fig = plt.figure(figsize=(20,20))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    for i in range(m):
        (data,inliers,outliers)=data_list[i]
        
        #plot the real line
        ax2=fig.add_subplot(3,m,i+1)
        X=np.linspace(-10,10,1000)
        Y=[a*x+b for x in X]
        model_label='Synth model with a = '+ str(a)+ " and b = "+str(b)
        ax2.plot(X,Y,color='green',label=model_label)
        
        plt.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7,label='inliers')
        if r_list[i]!=0:
            plt.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7,label='outliers')
        
        #Plot linearProgL1 line
        L1_line=linearProgL1(data)
        a_,b_=L1_line[0],L1_line[1]
        
        X=np.linspace(-10,10,1000)
        Y=[a_*x+b_ for x in X]
        irls_label='Linear Programmation using L1 norm with a = '+ str(a_)+ " and b = "+str(b_)
        ax2.plot(X,Y,color='blue',label=irls_label)
        ax2.title.set_text("Results for the ratio r = "+ str(int(r_list[i]*100)) + " %")
        ax2.legend(loc="lower center",bbox_to_anchor=(0.5, -0.5))
        plt.xlim([-10, 10])
        plt.ylim([-10, 10])
```

```python
linearProgL1_plot(data_list)
```

###### Linear Programming with $L_{\infty}$ norm

```python
def linearProgLinf(data):
    C=np.array([0,0,1]) # The vector considered here is a, b, u
    A_ub=np.zeros((2*N,3))
    b_ub=np.zeros(2*N)
    for i in range(N):
        A_ub[i,0]=-data[i,0]
        A_ub[i,1]=-1
        A_ub[i,2]=-1
        b_ub[i]=-data[i,1]
        
        A_ub[2*i+1,0]=data[i][0]
        A_ub[2*i+1,1]=1
        A_ub[2*i+1,2]=-1
        b_ub[2*i+1]=data[i,1]
    return linprog(C,A_ub=A_ub,b_ub=b_ub).x[:2]
    
```

```python
def linearProgLinf_plot(data_list):
    m=len(r_list)
    fig = plt.figure(figsize=(20,20))
    fig.subplots_adjust(hspace=0.3, wspace=0.3)
    for i in range(m):
        (data,inliers,outliers)=data_list[i]
        
        #plot the real line
        ax2=fig.add_subplot(3,m,i+1)
        X=np.linspace(-10,10,1000)
        Y=[a*x+b for x in X]
        model_label='Synth model with a = '+ str(a)+ " and b = "+str(b)
        ax2.plot(X,Y,color='green',label=model_label)
        
        plt.scatter(inliers[:,0],inliers[:,1],facecolor='white',edgecolor='blue',alpha=0.7,label='inliers')
        if r_list[i]!=0:
            plt.scatter(outliers[:,0],outliers[:,1],facecolor='white',edgecolor='red',alpha=0.7,label='outliers')
        
        #Plot linearProgL1 line
        L1_line=linearProgLinf(data)
        a_,b_=L1_line[0],L1_line[1]
        
        X=np.linspace(-10,10,1000)
        Y=[a_*x+b_ for x in X]
        irls_label='Linear Programmation using L infinity norm with a = '+ str(a_)+ " and b = "+str(b_)
        ax2.plot(X,Y,color='blue',label=irls_label)
        ax2.title.set_text("Results for the ratio r = "+ str(int(r_list[i]*100)) + " %")
        ax2.legend(loc="lower center",bbox_to_anchor=(0.5, -0.5))
        plt.xlim([-10, 10])
        plt.ylim([-10, 10])
```
